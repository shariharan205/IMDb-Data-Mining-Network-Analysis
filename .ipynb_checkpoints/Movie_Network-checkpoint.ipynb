<h2> Movie Network </h2>

```python
import pickle
import pandas as pd
import numpy as np
import operator
import math
from matplotlib import pyplot as plt
from statistics import mode
from collections import Counter

import networkx as nx
from networkx.algorithms import bipartite

from IPython.display import display
```

<h2> Data Preparation </h2>

```python
movie_genres = open("movie_genre.txt").read()
```

```python
movie_communities = open("communities.txt").read()
```

```python
communities = movie_communities.strip().split("\n")
```

<h2> Community Detection </h2>

```python
all_comm_movies = []
com_movies_list = []

for i in range(len(communities)):
    
    com_movies = communities[i].split("\t")
    com_movies_list.append(com_movies)
    all_comm_movies.extend(com_movies)
    
all_comm_movies = set(all_comm_movies)
```

```python
genre_dict = {}
genre_dict_all_comm = {}
movies = movie_genres.strip().split("\n")

for i in movies:
    j = i.split("\t\t")
    genre_dict[j[0]] = j[1]
    if j[0] in all_comm_movies:
        genre_dict_all_comm[j[0]] = j[1]
```

```python
def get_genre(movie_name):
    return genre_dict[movie_name]
```

<h2> Dominant Genre </h2>

```python
dominant_genres_qn7 = []
entire_dataset = []

i = 0
for com_movies in com_movies_list:

    genres = [genre_dict.get(j) for j in com_movies if genre_dict.get(j)]
    print("Community Number - ", i+1)
    dominant_genres_qn7.append(mode(genres))
    print("Most dominant genre - ", dominant_genres_qn7[-1])
    pd.Series(genres).value_counts().plot('bar')
    plt.show()
    i += 1
```

```python
most_dominant_genres = sorted(Counter(dominant_genres_qn7).items(), key=operator.itemgetter(1), reverse = True)
for i in most_dominant_genres:
    print(i[0], "-", i[1])
```

<h2> Most frequent across all movies irrespective of communities </h2>

```python
genre_dict_counter = Counter(genre_dict.values())
total_movies = sum(genre_dict_counter.values())
```

```python
print("\nInformation from Genres file only\n")
most_freq_genres = sorted(genre_dict_counter.items(), key=operator.itemgetter(1), reverse = True)
for i in most_freq_genres:
    print(i[0], "-", i[1])
```

```python
genre_dict_counter_all_comm = Counter(genre_dict_all_comm.values())

print("\nInformation from Network only\n")

most_freq_genres = sorted(genre_dict_counter_all_comm.items(), key=operator.itemgetter(1), reverse = True)
for i in most_freq_genres:
    print(i[0], "-", i[1])
```

<h2> Considering all movies as entire dataset</h2>

```python
dominant_genres_qn8b = []
for i in range(len(communities)):
    com_movies = communities[i].split("\t")
    genres = [genre_dict.get(i) for i in com_movies if genre_dict.get(i)]
    com_movies_len = len(genres) #number of movies in community with genre information
    genre_counts = Counter(genres)
    scores = {}
    
    for genre, c in genre_counts.items():
        p = c/com_movies_len
        q = genre_dict_counter[genre] / total_movies
        scores[genre] = math.log(c) * (p/q)

    dominant_genre = sorted(scores.items(), key=operator.itemgetter(1), reverse = True)[0]
    dominant_genres_qn8b.append(dominant_genre[0])

    
    print("Community Number - ", i+1)
    print("Most dominant genre - ", dominant_genre[0], " with score = ", dominant_genre[1])
    print("Most dominant genre according to previous question - ", dominant_genres_qn7[i])
    print("\n")
    print(sorted(scores.items(), key=operator.itemgetter(1), reverse = True))
    print("\n")
    
```

<h2> Considering movies across all communities as entire dataset </h2>

```python
genre_dict_counter_all_comm = Counter(genre_dict_all_comm.values())
total_movies_all_comm = sum(genre_dict_counter_all_comm.values())
```

```python
dominant_genres_qn8b = []
for i in range(len(communities)):
    com_movies = communities[i].split("\t")
    genres = [genre_dict.get(i) for i in com_movies if genre_dict.get(i)]
    com_movies_len = len(genres) #number of movies in community with genre information
    genre_counts = Counter(genres)
    scores = {}
    
    for genre, c in genre_counts.items():
        p = c/com_movies_len
        q = genre_dict_counter_all_comm[genre] / total_movies_all_comm
        scores[genre] = math.log(c) * (p/q)

    dominant_genre = sorted(scores.items(), key=operator.itemgetter(1), reverse = True)[0]
    dominant_genres_qn8b.append(dominant_genre[0])

    
    print("Community Number - ", i+1)
    print("Most dominant genre - ", dominant_genre[0], " with score = ", dominant_genre[1])
    print("Most dominant genre according to previous question - ", dominant_genres_qn7[i])
    print("\n")
    print(sorted(scores.items(), key=operator.itemgetter(1), reverse = True))
    print("\n")
```

<h2> Bipartite graph analysis </h2>

```python
with open('movie_actor_map.pkl', 'rb') as f:
    movie_actor_map = pickle.load(f)[0]
```

```python
with open('actor_movie_map.pkl', 'rb') as f:
    actor_movie_map = pickle.load(f)[0]
```

```python
def get_preferred_genre(actors):
    
    for actor in actors:
        movies_by_actor = actor_movie_map[actor]
        genres_by_actor = [genre_dict.get(i) for i in movies_by_actor if genre_dict.get(i)]
        print("Actor - ", actor, " - Genre counts of Actor - ", dict(Counter(genres_by_actor)))
```

```python
def get_labels(movies, actors, edges):
    
    labels_dict = {}
    
    i = 1
    keys_m, vals_m = [], []
    for movie in movies:
        labels_dict[movie] = 'M' + str(i)
        keys_m.append(movie)
        vals_m.append('M' + str(i))
        i = i + 1
        
    i = 1
    keys_a, vals_a = [], []
    for actor in actors:
        labels_dict[actor] = 'A' + str(i)
        keys_a.append(actor)
        vals_a.append('A' + str(i))
        i = i + 1
    
    labels = []

    for movie, actor in edges:
        labels.append((labels_dict[movie], labels_dict[actor]))

    return labels, [vals_m, keys_m], [vals_a, keys_a] 
        
```

```python
def plot_bipartite(community_number):
    
    print("Community number - ", community_number)
    
    movies_in_comm = communities[community_number-1].split('\t')
    
    print("\nDominant Genre by Q7 - ", dominant_genres_qn7[community_number-1])
    print("\nDominant Genre by Q8b - ", dominant_genres_qn8b[community_number-1])
    
    bipartite_edges = []
    bipartite_edges = [(movie, actor) for movie in movies_in_comm for actor in movie_actor_map[movie]]
    bipart_movies, bipart_actors = zip(*bipartite_edges)
    
    top3_actors, _ = zip(*Counter(bipart_actors).most_common(3))
    
    print("\nTop 3 most important actors - ", top3_actors)
    
    get_preferred_genre(top3_actors)
        
    
    B = nx.Graph()
    labels, legend_m, legend_a = get_labels(set(bipart_movies), set(bipart_actors), bipartite_edges)
    B.add_edges_from(labels)
    
    display(pd.DataFrame(legend_m).transpose())
    display(pd.DataFrame(legend_a).transpose())
    
    X, Y = bipartite.sets(B)
    pos = dict()
    pos.update( (n, (1, i)) for i, n in enumerate(X) ) # put nodes from X at x=1
    pos.update( (n, (2, i)) for i, n in enumerate(Y) ) # put nodes from Y at x=2
    nx.draw(B, pos=pos, with_labels=True, node_size=250, font_size=10, font_color='black', node_color='red', edge_color='yellow')
    plt.show() 
```

```python
plot_bipartite(25)
```

```python
plot_bipartite(28)
```
